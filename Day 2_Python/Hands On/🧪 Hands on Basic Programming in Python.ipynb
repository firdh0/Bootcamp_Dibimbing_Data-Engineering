# ✨ Attribute

```python
from datetime import datetime

class Mobil:
    def __init__(self, merek, tipe, harga, warna, tahun_pembuatan):
        self.merek  = merek
        self.tipe   = tipe
        self.harga  = harga
        self.warna  = warna
        self.tahun_pembuatan = tahun_pembuatan
        self.roda   = 4
        self.umur   = datetime.now().year - self.tahun_pembuatan
```

```python
nissan_march = Mobil("nissan", "march", 100_000_000, "putih", 2011)
```

```python
print("merek:", nissan_march.merek)
print("tipe:", nissan_march.tipe)
print("harga:", nissan_march.harga)
print("warna:", nissan_march.warna)
print("tahun_pembuatan:", nissan_march.tahun_pembuatan)
print("roda:", nissan_march.roda)
print("umur:", nissan_march.umur)
```

# ✨ Method

```python
from datetime import datetime

class Mobil:
    def __init__(self, merek, tipe, harga, warna, tahun_pembuatan):
        self.merek  = merek
        self.tipe   = tipe
        self.harga  = harga
        self.warna  = warna
        self.tahun_pembuatan = tahun_pembuatan
        self.roda   = 4
        self.umur   = datetime.now().year - self.tahun_pembuatan
        self.posisi = 0

    def klakson(self):
        print("tiiinnnn....")

    def maju(self):
        self.posisi += 1

    def mundur(self):
        self.posisi -= 1
```

```python
nissan_march = Mobil("nissan", "march", 100_000_000, "putih", 2011)
```

```python
nissan_march.klakson()
```

```python
print("posisi:", nissan_march.posisi)

nissan_march.maju()
print("posisi:", nissan_march.posisi)

nissan_march.mundur()
print("posisi:", nissan_march.posisi)

nissan_march.maju()
nissan_march.maju()
nissan_march.maju()
print("posisi:", nissan_march.posisi)

nissan_march.mundur()
nissan_march.mundur()
print("posisi:", nissan_march.posisi)
```

# ✨ Encapsulation

```python
class Warga:
    def __init__(self, nik):
        self.__nik = nik

galuh = Warga(1234567890)

print(galuh.__nik)
```

```python
class Warga:
    def __init__(self, nik):
        self.__nik = nik

    def get_nik(self):
        return self.__nik

galuh = Warga(1234567890)

print(galuh.get_nik())
```

```python
class Warga:
    def __init__(self):
        self.__nik = None

    def set_nik(self, nik):
        self.__nik = nik

    def get_nik(self):
        return self.__nik

galuh = Warga()

galuh.set_nik(1234567890)
print(galuh.get_nik())
```

# ✨ Inheritance

```python
class Warga:
    def menyapa(self):
        print("halo..")

class DataEngineer(Warga):
    pass

galuh = DataEngineer()
galuh.menyapa()
```

```python
class Warga:
    def __init__(self, nik):
        self.nik = nik

    def menyapa(self):
        print("halo..")

class DataEngineer(Warga):
    def __init__(self, nik):
        super().__init__(nik)

galuh = DataEngineer(1234567890)
print(galuh.nik)
```

# ✨ Multithreading

```python
import time
import random
import threading

def square(x):
    print("start:", x)
    time.sleep(random.randint(0, 10))
    print("finish:", x)
    return x * x

results = []
threads = []
for i in range(10):
    t = threading.Thread(target=lambda: results.append(square(i)))
    t.start()
    threads.append(t)

for t in threads:
    t.join()

print(results)
```

```python
import threading
import os

def pid(i):
    # printing process id
    print(f"\nID of process running worker {i}: {os.getpid()}")

# printing main program process id
print(f"ID of main process: {os.getpid()}")

threads = []
for i in range(2):
    t = threading.Thread(target=lambda: pid(i))
    t.start()
    threads.append(t)

for t in threads:
    t.join()
```

# ✨ Multiprocessing

```python
import time
import random
from multiprocessing import Pool

def square(x):
    print("start:", x)
    time.sleep(random.randint(0, 10))
    print("finish:", x)
    return x * x

with Pool(processes=4) as pool:
    result = pool.map(square, range(10))
print(result)
```

```python
import os
from multiprocessing import Pool

def pid(i):
    # printing process id
    print(f"\nID of process running worker {i}: {os.getpid()}")

# printing main program process id
print(f"ID of main process: {os.getpid()}")

with Pool(processes=4) as pool:
    result = pool.map(pid, range(2))
```

##  📬 String Formatting Methods

```python
name = "John"
age = 30

# Format string
message = "Halo, " + name + ". Anda berusia " + str(age) + " tahun."
print(message)

# Format string dengan f-strings
message = f"Halo, {name}. Anda berusia {age} tahun."
print(message)

# Format string dengan metode format
message = "Halo, {}. Anda berusia {} tahun.".format(name, age)
print(message)

# Format string dengan metode format dan parameter
message = "Halo, {name}. Anda berusia {age} tahun.".format(
        name = name,
        age = age,
    )
print(message)
```
